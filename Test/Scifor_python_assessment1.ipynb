```python
#14)Series object temp1,temp2,temp3,temp4 stores temprature of days of week 1,week2,week3,week4. Write python code to
#1. Print average temprature per week.
#2.Print average temperature of entire month.
# temperatures
temp1 = pd.Series([20, 25, 22, 23, 24], index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
temp2 = pd.Series([22, 24, 26, 27, 21], index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
temp3 = pd.Series([19, 21, 23, 25, 20], index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
temp4 = pd.Series([23, 24, 25, 22, 26], index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])

# average temperature per week
avg_temp_per_week = pd.DataFrame({
    'Week 1': temp1.mean(),
    'Week 2': temp2.mean(),
    'Week 3': temp3.mean(),
    'Week 4': temp4.mean()
}, index=['Average Temperature'])

print("Average temperature per week:")
print(avg_temp_per_week)

#average temperature of entire month
all_temps = pd.concat([temp1, temp2, temp3, temp4])
avg_temp_month = all_temps.mean()

print("\nAverage temperature of entire month:", avg_temp_month)
```
