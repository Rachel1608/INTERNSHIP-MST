```python
import seaborn as sns
```

```python
sns.set(style = 'dark')
data = {"X" : [11,30,67,80,100,116,125], "Y" : [100,300,500,700,800,900,1000], "Z":['Yes','No','No','Yes','No','Yes','Yes']}
sns.lineplot(x= data["X"], y = data['Y'],marker = '+', lw=1.5, ls='-')
```

```python
sns.barplot(x= data["X"], y = data['Y'], palette ='pastel')
```

```python
sns.boxplot(x= data["Z"], y = data['X'])
```

```python
sns.violinplot(x= data["Z"], y = data['Y'], hue=data['Z'])
```

```python
sns.distplot(x= data["Y"], kde =True, hist = True)
```
